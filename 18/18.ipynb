```python
array = [int(i) for i in open("17 (2).txt")]
q=[]
count=0
for n in range(len(array)):
    for k in range(n+1, len(array)):
        if (array[n]+array[k])%8==0:
            count+=1
            q.append(array[n]+array[k])
print(count, max(q))
```
