```python
from itertools import product # каждая буква используется любое количество раз
count = 0
for i in product("012345678", repeat = 6):
    s = ''.join(i)
    if s[0] != "0" and s.count("4") == 1 and sum([1 if int(x)%2 == 1 else 0 for x in s]) == 2:
        count += 1

print(count)
```

```python
from itertools import product
count=0
vowels = ["А", "Е"]
for i in product("АНДРЕЙ", repeat=4):
    
    s="".join(i)
    if s[0]!="Й" and sum([(x in vowels) for x in s])>=1:
        count+=1
print(count)
```
